```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv

# load the .env file variables
load_dotenv()

```

```python
client_id=os.environ.get('CLIENT_ID')
client_secret=os.environ.get('CLIENT_SECRET')
```

```python
client_secret
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

con = spotipy.Spotify(oauth_manager= SpotifyClientCredentials(client_id= client_id, client_secret= client_secret))
```

```python
id_artist = '00FQb4jTyendYWaN8pK0wa'

response = con.artist_top_tracks(id_artist)

print(response['tracks'])
```

```python
df_tracks = pd.DataFrame.from_records(response['tracks'])
print(df_tracks)
```

```python
diagrama_correlacion = sns.scatterplot(data=df_tracks, x='popularity', y='duration_ms')
fig = diagrama_correlacion.get_figure()
fig.savefig('scatter.png')
```
